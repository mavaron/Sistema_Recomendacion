```python
import graphlab as gl
from datetime import datetime
from os import path
import import_ipynb
import Conn_E_SQL as Conn

#Atributos de BD
host="recomdb.c4v5c6aergun.us-west-2.rds.amazonaws.com"
user="imaginex"
passwd="6^b{X2Du)Qf6"
database="RecomDB"
table= "CBF"

#Atributos de XML o consola

print ("Quieres ver apartamentos fuera de tu zona de escogencia?")
visualization=int(input())
Type=2

#Atributos de Vivienda

#Vector_BD_Vivienda= Conn.Conexion_BD(host,user,passwd,database,table)
#apartment_file_processed= gl.SFrame(C_V.Creacion_Vector_Apartamento(Vector_BD_Vivienda))
#apartment_file_normalized=gl.SFrame(Norm.norm_apartment_matrix(apartment_file_processed))
#Vector_IDF=[]
#Vector_IDF=Norm.Inv_DF(apartment_file_processed)

#Atributos de Cliente

clientfile="C:\\Users\\user\\Documents\\Sistema de Recomendacion\\Single_Client_raw.csv"
#client_file_processed=gl.SFrame(C_V.Creacion_Vector_Cliente(clientfile))
#client_array=client_file_processed["ID_C"]
#client_number=client_array[0]
#zona_array=client_file_processed["Zona"]
#zona_number=zona_array[0]

#Atributos de Clientes

Vector_Clientes_BD= Conn.Conexion_BD_clients(host,user,passwd,database,table)



#Atributos de Rating

#Vector_Ratings_BD=Conn.Conexion_BD_rating(client_number, host, user, passwd, database, table)
#client_rating=C_V.Creacion_Matrix_Rating(Vector_Ratings_BD)

#Atributos de Ratings

#Vector_Ratings_BD=Conn.Conexion_BD_ratings(host, user, passwd, database, table)
#clients_ratings=C_V.Creacion_Matrix_Rating(Vector_Ratings_BD)


def Creacion_Vector_Apartamento(Vector_BD_Vivienda):

        # Creacion de SFrames
        living_place_processed= gl.SFrame()
        
        #Arrays con Valores
        ID_V=[]
        Zona=[]
        Zonas_Comunes=[]
        Cercania_TM=[]
        Valorizacion=[]
        Cant_Viv_proy=[]
        Acceso_Discapacitados=[]
        Cuartos=[]
        Parqueaderos=[]
        Balcon=[]
        Mens_Cuota_Inicial=[]
        Cuota_Credito=[]
        k=0
        l=0
        Longitud_Vector=12

        # Asignacion de Valores BD a Vectores   
    
        while (l < Longitud_Vector):
            while (k< len(Vector_BD_Vivienda)) :
                ID_V.append(Vector_BD_Vivienda[k][l])
                Zona.append(Vector_BD_Vivienda[k][l+1])
                Zonas_Comunes.append(Vector_BD_Vivienda[k][l+2])
                Cercania_TM.append(Vector_BD_Vivienda[k][l+3])
                Valorizacion.append(Vector_BD_Vivienda[k][l+4])
                Cant_Viv_proy.append(Vector_BD_Vivienda[k][l+5])
                Acceso_Discapacitados.append(Vector_BD_Vivienda[k][l+6])
                Cuartos.append(Vector_BD_Vivienda[k][l+7])
                Parqueaderos.append(Vector_BD_Vivienda[k][l+8])
                Balcon.append(Vector_BD_Vivienda[k][l+9])
                Mens_Cuota_Inicial.append(Vector_BD_Vivienda[k][l+10])
                Cuota_Credito.append(Vector_BD_Vivienda[k][l+11])
                l=0
                k=k+1
            l=l+12             
                             
       
        #Edicion de SFrames de Apartamentos
        living_place_processed.add_column(gl.SArray(ID_V),name='ID_V')
        living_place_processed.add_column(gl.SArray(Zona),name='Zona')
        living_place_processed.add_column(gl.SArray(Zonas_Comunes),name='Zonas_Comunes')
        living_place_processed.add_column(gl.SArray(Cercania_TM),name='Cercania_TM')
        living_place_processed.add_column(gl.SArray(Valorizacion),name='Valorizacion')
        living_place_processed.add_column(gl.SArray(Cant_Viv_proy),name='Cant_Viv_Proy')
        living_place_processed.add_column(gl.SArray(Acceso_Discapacitados),name='Acceso_Discapacitados')
        living_place_processed.add_column(gl.SArray(Cuartos),name='Cuartos')
        living_place_processed.add_column(gl.SArray(Parqueaderos),name='Parqueaderos')
        living_place_processed.add_column(gl.SArray(Balcon),name='Balcon')
        living_place_processed.add_column(gl.SArray(Mens_Cuota_Inicial), name='Cuota_Inicial')
        living_place_processed.add_column(gl.SArray(Cuota_Credito), name='Cuota_Credito') 
                   
        
        return living_place_processed     

def Creacion_Vector_Cliente(clientfile):

        # Archivos CSV para Clientes 
        data_raw_client= clientfile

        # Creacion de SFrames
        client_raw= gl.SFrame.read_csv(data_raw_client,delimiter=';',header=True)
        clients_processed=gl.SFrame()

        #Variables

        Cant_personas=client_raw["Cant_Personas"]
        Estado_Civil=client_raw["Estado_Civil"]
        Destinacion_Vivienda=client_raw["Destinacion_Vivienda"]
        Edad=client_raw["Edad"]
        Edad_Hijo_Menor=client_raw["Edad_hijo_Menor"]
        Mascota=client_raw["Mascota"]
        Ocupacion=client_raw["Ocupacion"]
        Vehiculos=client_raw["Vehiculo"]
        Capacidad_Pago_Mes=client_raw["Capacidad_Pago_Mes"]
        Ahorros_Cuota_Inicial=client_raw["Ahorros_Cuota_Inicial"]
        Tiempo_Entrega=client_raw["Tiempo_Entrega"]
        Zonas_Comunes=[]
        Parqueaderos=[]
        Cercania_TM=[]
        Valorizacion=[]
        Cuartos=[]
        Balcon=[]
        Cant_Viv_proy=[]
        Cuota_Inicial=[]
        Capacidad_Pago_int=[]
        i=0
        j=0
        k=0
        l=0
        m=0
        n=0
        o=0
        p=0
        q=0
        r=0
        s=0

        #Edicion de SFrames de Clientes

        clients_processed.add_column(client_raw["ID_C"], name='ID_C')
        clients_processed.add_column(client_raw["Zona"], name='Zona')
        
       # Zonas Comunes

        while(j< Estado_Civil.size()):
                if (Edad[j] == 5 or Edad_Hijo_Menor[j]==1 or Edad_Hijo_Menor[j]==2 or Mascota[j]==1 or Estado_Civil[j]==1 ):
                    if (Estado_Civil[j]==1) :
                        Zonas_Comunes.append(1)
                    if Estado_Civil[j]!=1 :
                        Zonas_Comunes.append(3)
                else:
                    Zonas_Comunes.append(2)
                j=j+1
        clients_processed.add_column(gl.SArray(Zonas_Comunes), name='Zonas_Comunes')

         # Cercania TM

        while(l< Vehiculos.size()):
                if (Vehiculos[l]==0 or Vehiculos[l]==1):
                    if (Vehiculos[l]==0):
                        Cercania_TM.append(3)
                    else:
                        Cercania_TM.append(2)
                else:
                    Cercania_TM.append(1)
                l=l+1

        clients_processed.add_column(gl.SArray(Cercania_TM), name='Cercania_TM')
        
         # Valorizacion

        while (m< Destinacion_Vivienda.size()):
            if (Destinacion_Vivienda[m]==0):
                Valorizacion.append(1)
            else:
                Valorizacion.append(3)
            m=m+1
        clients_processed.add_column(gl.SArray(Valorizacion), name='Valorizacion')
        
        
        #Cantidad_Vivendas_Proyecto

        while(o<Ocupacion.size()):
            if (Ocupacion[o] == 4 or Edad_Hijo_Menor[o]==1 or Edad_Hijo_Menor[o]==2):
                if(Ocupacion[o]==4):
                    Cant_Viv_proy.append(1)
                else:
                    Cant_Viv_proy.append(3)
            else:
                Cant_Viv_proy.append(2)
            o=o+1
        clients_processed.add_column(gl.SArray(Cant_Viv_proy), name='Cant_Viv_proy') 

        # Acceso_Discapacitados
        
        clients_processed.add_column(client_raw["Discapacidad"], name='Acceso_Discapacitados')
        
        # Tratamiento_Cuartos

        while(i< Estado_Civil.size()):
                if Estado_Civil[i] == 1 or  Estado_Civil[i] == 4:
                    Cuartos.append(Cant_personas[i])
                else:
                    Cuartos.append(Cant_personas[i]-1)
                i=i+1
        clients_processed.add_column(gl.SArray(Cuartos), name='Cuartos')
        
                
        # Parqueaderos

        while(k< Vehiculos.size()):
                if (Vehiculos[k]==0):
                    Parqueaderos.append(0)
                if (Vehiculos[k]==1):
                    Parqueaderos.append(1)
                if (Vehiculos[k]==2):
                    Parqueaderos.append(2)
                if (Vehiculos[k]>=3):
                    Parqueaderos.append(3)
                k=k+1
        clients_processed.add_column(gl.SArray(Parqueaderos), name='Parqueaderos')            

        # Balcon

        while (n< Ocupacion.size()):
             if (Edad[n] == 1 or Estado_Civil[n]==1 or Estado_Civil[n]==2 or Ocupacion[n]==1 or  Mascota[n]==1):
                Balcon.append(1)
             else:
                Balcon.append(0)
             n=n+1
        clients_processed.add_column(gl.SArray(Balcon), name='Balcon')

        #Cuota_Inicial

        while(p<Capacidad_Pago_Mes.size()):
            if str(Capacidad_Pago_Mes[p]).find(",") == -1:
                Capacidad_Pago_int.append(float(Capacidad_Pago_Mes[p]))
            else:
                q=Capacidad_Pago_Mes[p].replace(',','.')
                Capacidad_Pago_int.append(float(q))
            p=p+1 


        while (r<Capacidad_Pago_Mes.size()):
            Cuota_Inicial.append(Capacidad_Pago_int[r]*Tiempo_Entrega[r]+Ahorros_Cuota_Inicial[r])
            r=r+1

        clients_processed.add_column(gl.SArray(Cuota_Inicial), name='Cuota_Inicial')

        #Cuota Credito

        clients_processed.add_column(gl.SArray(Capacidad_Pago_int), name='Cuota_Credito')


        return clients_processed

def Creacion_BD_Clientes(Vector_Clientes_BD):

        #Creacion de SFrames
        clients_processed=gl.SFrame()

        #Variables

        ID_C=[]
        Zona=[]
        Cant_personas=[]
        Estado_Civil=[]
        Destinacion_Vivienda=[]
        Edad=[]
        Edad_Hijo_Menor=[]
        Mascota=[]
        Discapacidad=[]
        Ocupacion=[]
        Vehiculos=[]
        Capacidad_Pago_Mes=[]
        Ahorros_Cuota_Inicial=[]
        Tiempo_Entrega=[]
        Zonas_Comunes=[]
        Parqueaderos=[]
        Cercania_TM=[]
        Valorizacion=[]
        Cuartos=[]
        Balcon=[]
        Cant_Viv_proy=[]
        Cuota_Inicial=[]
        
        i=0
        j=0
        k=0
        l=0
        m=0
        n=0
        o=0
        p=0
        q=0
        r=0
        s=0
        t=0
        u=0
        Longitud_Vector=14

        # Asignacion de Valores BD a Vectores   
    
        while (u < Longitud_Vector):
            while (t< len(Vector_Clientes_BD)) :
                ID_C.append(Vector_Clientes_BD[t][u])
                Zona.append(Vector_Clientes_BD[t][u+1])
                Edad.append(Vector_Clientes_BD[t][u+2])
                Estado_Civil.append(Vector_Clientes_BD[t][u+3])
                Ocupacion.append(Vector_Clientes_BD[t][u+4])
                Cant_personas.append(Vector_Clientes_BD[t][u+5])
                Edad_Hijo_Menor.append(Vector_Clientes_BD[t][u+6])
                Discapacidad.append(Vector_Clientes_BD[t][u+7])
                Mascota.append(Vector_Clientes_BD[t][u+8])
                Vehiculos.append(Vector_Clientes_BD[t][u+9])
                Destinacion_Vivienda.append(Vector_Clientes_BD[t][u+10])
                Capacidad_Pago_Mes.append(Vector_Clientes_BD[t][u+11])
                Ahorros_Cuota_Inicial.append(Vector_Clientes_BD[t][u+12])
                Tiempo_Entrega.append(Vector_Clientes_BD[t][u+13])
                u=0
                t=t+1
            u=u+14             

        #Edicion de SFrames de Clientes

        clients_processed.add_column(gl.SArray(ID_C), name='ID_C')
        clients_processed.add_column(gl.SArray(Zona), name='Zona')
        
       # Zonas Comunes

        while(j< len(Estado_Civil)):
                if (Edad[j] == 5 or Edad_Hijo_Menor[j]==1 or Edad_Hijo_Menor[j]==2 or Mascota[j]==1 or Estado_Civil[j]==1 ):
                    if (Estado_Civil[j]==1) :
                        Zonas_Comunes.append(1)
                    if Estado_Civil[j]!=1 :
                        Zonas_Comunes.append(3)
                else:
                    Zonas_Comunes.append(2)
                j=j+1
        clients_processed.add_column(gl.SArray(Zonas_Comunes), name='Zonas_Comunes')

         # Cercania TM

        while(l< len(Vehiculos)):
                if (Vehiculos[l]==0 or Vehiculos[l]==1):
                    if (Vehiculos[l]==0):
                        Cercania_TM.append(3)
                    else:
                        Cercania_TM.append(2)
                else:
                    Cercania_TM.append(1)
                l=l+1

        clients_processed.add_column(gl.SArray(Cercania_TM), name='Cercania_TM')
        
         # Valorizacion

        while (m< len(Destinacion_Vivienda)):
            if (Destinacion_Vivienda[m]==0):
                Valorizacion.append(1)
            else:
                Valorizacion.append(3)
            m=m+1
        clients_processed.add_column(gl.SArray(Valorizacion), name='Valorizacion')
        
        
        #Cantidad_Vivendas_Proyecto

        while(o<len(Ocupacion)):
            if (Ocupacion[o] == 4 or Edad_Hijo_Menor[o]==1 or Edad_Hijo_Menor[o]==2):
                if(Ocupacion[o]==4):
                    Cant_Viv_proy.append(1)
                else:
                    Cant_Viv_proy.append(3)
            else:
                Cant_Viv_proy.append(2)
            o=o+1
        clients_processed.add_column(gl.SArray(Cant_Viv_proy), name='Cant_Viv_proy') 

        # Acceso_Discapacitados
        
        clients_processed.add_column(gl.SArray(Discapacidad), name='Acceso_Discapacitados')
        
        # Tratamiento_Cuartos

        while(i< len(Estado_Civil)):
                if Estado_Civil[i] == 1 or  Estado_Civil[i] == 4:
                    Cuartos.append(Cant_personas[i])
                else:
                    Cuartos.append(Cant_personas[i]-1)
                i=i+1
        clients_processed.add_column(gl.SArray(Cuartos), name='Cuartos')
        
              
        # Parqueaderos

        while(k< len(Vehiculos)):
                if (Vehiculos[k]==0):
                    Parqueaderos.append(0)
                if (Vehiculos[k]==1):
                    Parqueaderos.append(1)
                if (Vehiculos[k]==2):
                    Parqueaderos.append(2)
                if (Vehiculos[k]>=3):
                    Parqueaderos.append(3)
                k=k+1
        clients_processed.add_column(gl.SArray(Parqueaderos), name='Parqueaderos')            

        # Balcon

        while (n< len(Ocupacion)):
             if (Edad[n] == 1 or Estado_Civil[n]==1 or Estado_Civil[n]==2 or Ocupacion[n]==1 or  Mascota[n]==1):
                Balcon.append(1)
             else:
                Balcon.append(0)
             n=n+1
        clients_processed.add_column(gl.SArray(Balcon), name='Balcon')

        #Cuota Inicial
        
        while (r<len(Capacidad_Pago_Mes)):
            Cuota_Inicial.append(Capacidad_Pago_Mes[r]*Tiempo_Entrega[r]+Ahorros_Cuota_Inicial[r])
            r=r+1

        clients_processed.add_column(gl.SArray(Cuota_Inicial), name='Cuota_Inicial')

        #Cuota Credito

        clients_processed.add_column(gl.SArray(Capacidad_Pago_Mes), name='Cuota_Credito')


        return clients_processed
    
  
    
def Creacion_Matrix_Rating(Vector_Ratings_BD):
        
        # Creacion de SFrames
        rating_processed= gl.SFrame()
        
        #Arrays con Valores
        ID_C=[]
        ID_V=[]
        Rating=[]
        Longitud_Vector=3
        l=0
        k=0

        # Asignacion de Valores a Vectores   
    
        while (l < Longitud_Vector):
            while (k< len(Vector_Ratings_BD)) :
                ID_C.append(Vector_Ratings_BD[k][l])
                ID_V.append(Vector_Ratings_BD[k][l+1])
                Rating.append(Vector_Ratings_BD[k][l+2])
                l=0
                k=k+1
            l=l+4             
                             
       
        #Edicion de SFrames de Rating
        rating_processed.add_column(gl.SArray(ID_C),name='ID_C')
        rating_processed.add_column(gl.SArray(ID_V),name='ID_V')
        rating_processed.add_column(gl.SArray(Rating),name='Rating')
        
        return rating_processed


```
